```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_source_path = "DataTarget/"
```

## Data Per Vendor Analysis

```python
df_per_vendor = pd.read_parquet(data_source_path+"df_per_vendor")
```

```python
df_per_vendor.columns
```

```python
df_per_vendor = df_per_vendor.set_index("vendor_name")
```

### Count vs Passenger Count

```python
# Total count for records in the Data
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="count")\
.set(title='Total records per vendor')

# As per below chart, it is seen we have least no of DDS count in Data
```

```python
# Total passenger count for every vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Passenger_Count_sum")\
.set(title='Total passenger count per vendor')

# As per below chart, it is seen total passenger count is higher for VTS
```

```python
# Avergae passenger count for every vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Passenger_Count_avg")\
.set(title='Avergae Passenger count per vendor')

# Also average passenger cnt is higher for VTS is higher
# Based on count and Passsenger count data, it is seen VTS have relatively higher average passenger cnt per trip
```

From above charts, I could see the CMT and VTS have toe-to-toe TAXI Ride counts, but VTS has better average Passenger count

### Vendor Vs Cost

```python
# Total Fare per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Fare_Amt_sum")\
.set(title='Total Fare Amt per vendor')

# Even though the passenger count is high in VTS, the total fare amount is not much higher w.r.t to CMT
```

```python
# Average Fare per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Fare_Amt_avg")\
.set(title='Avg Fare Amt per vendor')

# But the average Fare amount is almost same fot all the three
# But in VTS, the avergae passenger count is higher, for individual passenger the cost is half
```

From above two charts I could see, eventhough VTS has higher Avg Passenger count, their cost is almost same as CMT.

Which means cost per passenger is lower for VTS

```python
# Total Surcharge per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="surcharge_sum")\
.set(title='Total Surcharge Amt per vendor')

# Surprisingly the CMT have zero surcharge amount
# Other two vendor is in line with count data
```

```python
# Avg Surcharge per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="surcharge_avg")\
.set(title='Avg Surcharge Amt per vendor')

# The Average surcharge is almost same bt VTS and DDS
```

From above chart, I could see CMT has zero surcharge, while VTS and DDS has toe-to-toe avg surcharge amt

```python
# Total Tip Amt per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Tip_Amt_sum")\
.set(title='Total Tip_Amt Amt per vendor')

# Tip Amt is in align with count
```

```python
# Avg Tip Amt per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Tip_Amt_avg")\
.set(title='Avg Tip_Amt Amt per vendor')

# Seems like Tip amount in VTS is relatively higher than others
# Even the VTS count is also higher which give me assumption that VTS provides good service
```

We could see VTS has better Tip_Amt than other vendors. 

We also saw avg passenger cnt is higher in VTS which gives us assumption whether passenger count causes higher avg Tip_Amt

```python
# Total Tolls_Amt Amt per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Tolls_Amt_sum")\
.set(title='Total Tolls_Amt Amt per vendor')

# Aligns with count records
```

```python
# Avg Tolls_Amt Amt per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Tolls_Amt_avg")\
.set(title='Avg Tolls_Amt Amt per vendor')

# Tolls amt is higher in DDS
# We early seen that Avg Trip Distance is slightly higher in DDS which justifies higher Tolls amt

# Does this mean DDS is preffered for longer distance??
```

Tolls amt is higher in DDS

Does DDS cover longer distance?

```python
# Avg Tolls_Amt Amt per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Total_Amt_sum")\
.set(title='Total Total_Amt Amt per vendor')

# ALigns with count
```

```python
# Avg Total_Amt Amt per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Total_Amt_avg")\
.set(title='Avg Total_Amt Amt per vendor')

# On checking Total Amt all vendor are almost toe-to-toe
```

Inspite of varying surcharges, tolls, tips, On checking Total Amt all vendor are almost toe-to-toe

```python
# Total Fare distribution in percentage terms
df_per_vendor[[
    'Fare_Amt_avg', 
    'surcharge_avg', 
    'Tip_Amt_avg',
    'Tolls_Amt_avg', 
    'Total_Amt_avg']].div(df_per_vendor["Total_Amt_avg"], axis=0) * 100
```

```python
df_per_vendor[[
    'Fare_Amt_avg', 
    'surcharge_avg', 
    'Tip_Amt_avg',
    'Tolls_Amt_avg', 
    'Total_Amt_avg']]

# For CMT I could see the surcharge is being replaced by Fare Amount
```

The above two Cost distribution gives a clear pic on how Total fare isdistributed for each vendor.

Notably in eventhough CMT has 0 surcharges, its Fare amount is higher which neglect surcharge advantage

### Duration Vs Distance

```python
# Total Distance covered per vendor
sns.barplot(data=df_per_vendor, x=df_per_vendor.index, y="Trip_Distance_sum")\
.set(title='Total Distance covered per vendor')

# The Total Distance convered is aligns with total_count
```

## Data Per Payment

```python
df_per_payment_type = pd.read_parquet(data_source_path+"df_per_payment_type")
```

```python
df_per_payment_type = df_per_payment_type.set_index("Payment_Type")
```

```python
df_per_payment_type.columns
```

```python
df_per_payment_type
```

### Count Vs Payment

```python
# Total count
sns.barplot(data=df_per_payment_type, x=df_per_payment_type.index, y="count")\
.set(title="Count per PaymentType")
```

As per above Chart It is seen that Most of te transaction are done by CASH

While there is a considerable CREDIT 

With very minimal NO CHARGE and DISPUTE

### Payment Vs Cost Distribution

```python
# Total Fare per Payment
sns.barplot(data=df_per_payment_type, x=df_per_payment_type.index, y="Fare_Amt_sum")\
.set(title='Total Fare Amt per Payment')

# Aligns with Count
```

```python
# Average Fare per Payement
sns.barplot(data=df_per_payment_type, x=df_per_payment_type.index, y="Fare_Amt_avg")\
.set(title='Avg Fare Amt per PaymentType')

# On checking Average, it seen that avg cost is higher for DISPUTE and CREDIT.
# CASH settlment has least average
```

While cause of max count, CASH has highest total fare 

But when considered its mean, it is seen that CASH settlement has average fare mean

While DISPUTE and CREDIT have higher mean fare.

Does Higher Cost and DISPUT,CREDIT are related?

It gives me a assumption that when Fare is less it leads to smooth settlement in CASH but when it raises it seems like leading to DISPUTE and CREDIT

```python
# Avg Surcharge per Payment
sns.barplot(data=df_per_payment_type, x=df_per_payment_type.index, y="surcharge_avg")\
.set(title='Avg Surcharge Amt per PaymentType')

# The Average surcharge is almost same bt VTS and DDS
```

It is seen that with DISPUTE and NO_CHARGE the surcharge colected is close to Zero

While with CASH and CREDIT it averages to .175

```python
# Total Tip Amt per PaymentType
sns.barplot(data=df_per_payment_type, x=df_per_payment_type.index, y="Tip_Amt_sum")\
.set(title='Total Tip_Amt Amt per PaymentType')

# Surprisingly CREDIT transaction leads to higher total tip amt collected
```

```python
# Avg Tip Amt per PaymentType
sns.barplot(data=df_per_payment_type, x=df_per_payment_type.index, y="Tip_Amt_avg")\
.set(title='Avg Tip_Amt Amt per PaymentType')

# It also seems like CREDIT transaction leads to higher average Tip amount
```

It seems like when Cutomer pay via CREDIT they tend to pay Higher Tip amount
Also we seens earlier Fare amount for CREDIT is higher than CASH so Tip might be offered relative to Fare

And also normally confluent people used to have CREDIT which seems related

```python
# Avg Tolls_Amt per PaymentType
sns.barplot(data=df_per_payment_type, x=df_per_payment_type.index, y="Tolls_Amt_avg")\
.set(title='Avg Tolls_Amt Amt per PaymentType')
```

CASH and NO_CHARGE has lower Toll charge

While DISPUTE has highest

```python
# Avg Total_Amt Amt per PaymentType
sns.barplot(data=df_per_payment_type, x=df_per_payment_type.index, y="Total_Amt_avg")\
.set(title='Avg Total_Amt Amt per PaymentType')

# On checking Total Amt all PaymentType are almost toe-to-toe
```

On aveage, the Total amount in CREDIT is higher even though its Fare is lower compared to DISPUTE.

Which gives me a assumption that higher Tips might contribute to higher charge

```python
df_per_payment_type[[
    'Fare_Amt_avg', 
    'surcharge_avg', 
    'Tip_Amt_avg',
    'Tolls_Amt_avg', 
    'Total_Amt_avg']]

# For CMT I could see the surcharge is being replaced by Fare Amount
```
